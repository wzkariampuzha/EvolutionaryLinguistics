```python
amer_directory = '../Ngrams/amer_unigram_data/'
brit_directory = '../Ngrams/brit_unigram_data/'

def open_json(directory,file_path):
    with open(directory+file_path,'r') as f:
        json_dictionary = json.load(f)
        f.close()
    return json_dictionary
```

1800-1850 Set Analysis

```python
BRIT_1800_1850_STEP3 = open_json(brit_directory,'LEXICON_1800-1850_STEP3.json')
AMER_1800_1850_STEP3 = open_json(amer_directory,'LEXICON_1800-1850_STEP3.json')

brit_1800_1850_s3 = set(BRIT_1800_1850_STEP3.keys())
amer_1800_1850_s3 = set(AMER_1800_1850_STEP3.keys())

intersection_1800_1850 = brit_1800_1850_s3.intersection(amer_1800_1850_s3)
amer_unique_1800_1850_s3 = amer_1800_1850_s3.difference(intersection_1800_1850)
brit_unique_1800_1850_s3 = brit_1800_1850_s3.difference(intersection_1800_1850)
```
